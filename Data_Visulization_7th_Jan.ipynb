```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = ["Jan", "Feb", "March", "April", "May", "June"]
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]

# Create a histogram for energy consumption
# Using np.arange to generate numerical bin edges for the histogram
#the range of energy consumption
min_consumption = min(energy_consumption)
max_consumption = max(energy_consumption)

# Calculate the number of bins
num_bins = len(energy_consumption)
bin_edges = np.linspace(min_consumption, max_consumption, num_bins + 1)

# Create the histogram with numerical bins
plt.hist(energy_consumption, bins=bin_edges, edgecolor='black')

# Set labels and title
plt.title("Energy Consumption Over 6 Months")
plt.xlabel("Energy Consumption (MWh)")
plt.ylabel("Frequency")

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data1 = {
    "Solar": [1200,1500,2000],
    "Wind": [3000,3200,3400],
    "Hydropower": [1700,2000,1900],
    "Biomass":[2500,3001,2700]
}
df =  pd.DataFrame(data1)
sns.pairplot(df)
plt.show()
```

```python
#Heat Map
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',linewidth=0.5)
plt.title('Correlation Energy Sources')
plt.show()
```

```python
#create a boxplot
sns.boxplot(data=df)
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
#create a violin plot
sns.violinplot(data=df)
plt.title('Violin plot of Energy Consumption Distribution')
plt.show()
```

```python
#Sample data for energy consumption and carbon emission
energy_values = [100,200,300,400,500]
carbon_values = [10,20,30,40,50]
df_reg = pd.DataFrame({'Energy Consumption': energy_values, 'Carbon Emission': carbon_values})
sns.regplot(x='Energy Consumption', y='Carbon Emission', data=df_reg)
plt.title('Relationship between Energy Consumption and Carbon Emission')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the IRIS dataset (assuming it's in a file named 'IRIS.csv' in your Colab environment)
try:
  iris = pd.read_csv('IRIS.csv')
except FileNotFoundError:
  print("Error: 'IRIS.csv' not found. Please upload the file to your Colab environment.")
  exit()
# Ensure the dataset is loaded correctly
if iris.empty:
  print("Error: IRIS dataset is empty.")
  exit()
# Display the first few rows to verify it's loaded
print(iris.head())

# 1. Pair Plot
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of IRIS Dataset Features', y=1.02)
plt.show()

# 2. Heatmap of Correlation Matrix
correlation_matrix = iris.drop('species', axis=1).corr()  # Exclude 'species' for correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of IRIS Features')
plt.show()

# 3. Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris.drop('species', axis=1))  # Exclude 'species' for boxplot
plt.title('Boxplot of IRIS Features')
plt.show()

# 4. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris.drop('species', axis=1)) # Exclude 'species' for violin plot
plt.title('Violin Plot of IRIS Features')
plt.show()

# 5. Scatter plot of Petal length and Petal width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species')
plt.title('Scatter Plot of Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()

# 6.  Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x='sepal_length', hue='species', element='step')
plt.title('Histogram of Sepal Length')
plt.show()
```
